```python
import pandas as pd

data = {
    "Project": [
        "Climeworks Orca",
        "Climeworks Mammoth",
        "Carbon Engineering Pilot",
        "DOE Long-Term Target"
    ],
    "Technology": [
        "Solid DAC",
        "Solid DAC",
        "Liquid DAC",
        "Future DAC Goal"
    ],
    "Cost_USD_per_tCO2": [
        1000,
        700,
        400,
        100
    ]
}

df = pd.DataFrame(data)

df
```

```python
print(df.info())
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    df["Project"],
    df["Cost_USD_per_tCO2"]
)

plt.title("Cost Comparison of Direct Air Capture Projects")

plt.xlabel("Project")

plt.ylabel("Cost (USD per tCO₂)")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(12,6))

bars = plt.bar(
    df["Project"],
    df["Cost_USD_per_tCO2"]
)

plt.title(
    "Direct Air Capture Cost Comparison",
    fontsize=16
)

plt.xlabel(
    "Projects",
    fontsize=12
)

plt.ylabel(
    "Cost ($/tCO₂)",
    fontsize=12
)

plt.grid(axis='y')

plt.xticks(rotation=15)

for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 20,
        f"${height}",
        ha='center'
    )

plt.tight_layout()

plt.show()
```

```python
import pandas as pd

df_scale = pd.DataFrame({
    "CO2_Removed_Gt": [1, 5, 10],
    "Cost_per_ton_USD": [200, 200, 200]
})

df_scale["Total_Cost_Billion_USD"] = (
    df_scale["CO2_Removed_Gt"] * 1000 * df_scale["Cost_per_ton_USD"] / 1000
)

df_scale
```

```python
import pandas as pd

co2_scenarios = [1, 5, 10]
cost_per_ton = 200

df_scale = pd.DataFrame({
    "CO2_Removed_Gt": co2_scenarios
})

df_scale["Cost_per_ton_USD"] = cost_per_ton

df_scale["Total_Cost_Billion_USD"] = (
    df_scale["CO2_Removed_Gt"] * cost_per_ton
)

df_scale
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

bars = plt.bar(
    df_scale["CO2_Removed_Gt"].astype(str),
    df_scale["Total_Cost_Billion_USD"]
)

plt.title("Annual DAC Cost at $200 per ton CO₂")
plt.xlabel("CO₂ Removed (Gt/year)")
plt.ylabel("Annual Cost (Billion USD)")

for bar in bars:
    y = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        y + 20,
        f"${y:.0f}B",
        ha='center'
    )

plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
import pandas as pd

df_compare = pd.DataFrame({
    "Metric": [
        "Steel",
        "Cement",
        "Aluminum",
        "Land"
    ],
    
    "Liquid_DAC": [
        50000,
        20000,
        0,
        100
    ],
    
    "Solid_DAC": [
        40000,
        20000,
        10000,
        220
    ]
})

df_compare
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(df_compare["Metric"]))
width = 0.35

plt.figure(figsize=(10,6))

plt.bar(
    x - width/2,
    df_compare["Liquid_DAC"],
    width,
    label="Liquid DAC"
)

plt.bar(
    x + width/2,
    df_compare["Solid_DAC"],
    width,
    label="Solid DAC"
)

plt.xticks(
    x,
    df_compare["Metric"]
)

plt.title(
    "Liquid DAC vs Solid DAC Resource Requirements"
)

plt.ylabel("Amount")

plt.legend()

plt.grid(axis='y', alpha=0.3)

plt.show()
```
